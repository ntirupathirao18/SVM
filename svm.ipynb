```python
import pandas as pd 
import os
import numpy as np
```

```python
filename = os.path.abspath('voice.csv')
print(filename)
df = pd.read_csv(filename)
print(type(df['label'][1]))
```

```python
def change_data(data,col1) :
    data[col1].replace(['female','male'],[0,1],inplace=True)  
    return data
```

```python
def removecolumns(pandasArray, *column):
    return pandasArray.drop(pandasArray.columns[[column]], axis=1).values
```

```python
data = change_data(df,'label')
print(data)
np_array = removecolumns(data,)
target = np.array(data['label'])
print(len(target))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data,target,test_size = 0.2, random_state = 45)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train,y_train)  
```

```python
predt = clf.predict(x_test)

from sklearn import metrics
print('accurancy :',metrics.accuracy_score(y_test,predt))
```

```python
clf_linearsvc = svm.LinearSVC()
clf_linearsvc.fit(x_train,y_train)
```

```python
predt_linear = clf_linearsvc.predict(x_test)
```

```python
from sklearn import metrics
print('accurancy :',metrics.accuracy_score(y_test,predt_linear))
```
